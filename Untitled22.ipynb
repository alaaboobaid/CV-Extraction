# CV Information Extraction & ATS Scoring with Internship Matching

```python
import pdfplumber
import spacy
import re
import phonenumbers
import json
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List
from sentence_transformers import SentenceTransformer, util
from reportlab.pdfgen import canvas  # Only used to create the dummy PDF for testing

# ==========================================
# 1. THE CV EXTRACTOR CLASS (ATS SYSTEM)
# ==========================================
class CVExtractor:
    """
    Extracts text from PDF, parses it into structured data,
    prints the result, and saves to JSON.
    """
    def __init__(self):
        print("‚öôÔ∏è  Initializing CV Extractor (Loading NLP models)...")
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except:
            from spacy.cli import download
            download("en_core_web_sm")
            self.nlp = spacy.load("en_core_web_sm")

    def process_cv(self, pdf_path: str) -> Dict:
        """
        Main function to Extract -> Print -> Save
        """
        print(f"\nüìÇ Reading PDF: {pdf_path}...")
        raw_text = self._extract_text_from_pdf(pdf_path)

        # Parse data
        extracted_data = {
            "personal_info": self._extract_personal_info(raw_text),
            "education": self._extract_section(raw_text, ["EDUCATION", "ACADEMIC BACKGROUND"]),
            "experience": self._extract_experience_robust(raw_text),
            "skills": self._extract_skills(raw_text),
            "raw_text": raw_text[:500] + "..." # Store snippet only to keep JSON clean
        }

        # 1. PRINT RESULTS
        self._print_extraction_results(extracted_data)

        # 2. SAVE TO JSON
        json_filename = pdf_path.replace(".pdf", "_extracted.json")
        with open(json_filename, 'w', encoding='utf-8') as f:
            json.dump(extracted_data, f, indent=4)
        print(f"\nüíæ Data successfully saved to: {json_filename}")

        return extracted_data

    def _extract_text_from_pdf(self, pdf_path: str) -> str:
        text = ""
        try:
            with pdfplumber.open(pdf_path) as pdf:
                for page in pdf.pages:
                    text += (page.extract_text() or "") + "\n"
        except Exception as e:
            print(f"‚ùå Error reading PDF: {e}")
        return text

    def _extract_personal_info(self, text: str) -> Dict:
        info = {"name": "Not Found", "email": "Not Found", "phone": "Not Found", "website": "Not Found"}

        # Extract Email
        email = re.search(r'[\w\.-]+@[\w\.-]+\.\w+', text)
        if email: info['email'] = email.group(0)

        # Extract Phone (Palestinian format +970/972)
        try:
            for match in phonenumbers.PhoneNumberMatcher(text, "PS"):
                info["phone"] = phonenumbers.format_number(match.number, phonenumbers.PhoneNumberFormat.INTERNATIONAL)
                break
        except: pass

        # Extract Website
        url = re.search(r'https?://[^\s]+', text)
        if url: info['website'] = url.group(0)

        # Extract Name (Heuristic: First non-empty line)
        lines = [l.strip() for l in text.split('\n') if l.strip()]
        if lines: info['name'] = lines[0]

        return info

    def _extract_section(self, text: str, headers: List[str]) -> List[str]:
        """Generic section extractor"""
        lines = text.split('\n')
        content = []
        capture = False

        for line in lines:
            line = line.strip()
            # Start capturing if header matches
            if any(h in line.upper() for h in headers):
                capture = True
                continue

            # Stop capturing if new section starts (common headers)
            common_headers = ['EXPERIENCE', 'EDUCATION', 'SKILLS', 'PROJECTS', 'SUMMARY', 'LANGUAGES']
            if capture and any(h in line.upper() for h in common_headers) and len(line) < 20:
                capture = False

            if capture and line:
                content.append(line)
        return content

    def _extract_experience_robust(self, text: str) -> List[str]:
        """
        Improved Experience Logic:
        1. Extract the experience block.
        2. Use Regex to find dates (years) to identify distinct entries.
        """
        raw_lines = self._extract_section(text, ["EXPERIENCE", "WORK HISTORY", "PROFESSIONAL EXPERIENCE"])
        if not raw_lines: return []

        entries = []
        current_entry = ""

        # Regex for dates like "2021", "2016 - 2018", "Sep 2021"
        date_pattern = re.compile(r'(19|20)\d{2}')

        for line in raw_lines:
            # If a line contains a date and we already have text, push the previous entry
            if date_pattern.search(line) and len(current_entry) > 10:
                entries.append(current_entry.strip())
                current_entry = line # Start new entry
            else:
                current_entry += " " + line

        if current_entry: entries.append(current_entry.strip())

        return entries

    def _extract_skills(self, text: str) -> List[str]:
        raw_lines = self._extract_section(text, ["SKILLS", "TECHNICAL SKILLS"])
        full_text = " ".join(raw_lines)
        # Split by bullets, commas, or newlines
        skills = re.split(r'[,‚Ä¢\n]', full_text)
        # Clean and deduplicate
        return list(set([s.strip() for s in skills if len(s.strip()) > 1]))

    def _print_extraction_results(self, data: Dict):
        print("\n" + "="*50)
        print("üìÑ EXTRACTED CV DATA")
        print("="*50)
        print(f"üë§ Name:    {data['personal_info']['name']}")
        print(f"üìß Email:   {data['personal_info']['email']}")
        print(f"üì± Phone:   {data['personal_info']['phone']}")
        print(f"üåê Website: {data['personal_info']['website']}")

        print("\nüéì Education:")
        for edu in data['education']: print(f"  - {edu}")

        print("\nüíº Experience:")
        for exp in data['experience']: print(f"  - {exp}")

        print("\nüõ†Ô∏è  Skills:")
        print(f"  {', '.join(data['skills'])}")
        print("="*50)


# ==========================================
# 2. THE MATCHING SYSTEM (AI & ATS SCORING)
# ==========================================
class InternshipMatchingSystem:
    def __init__(self, csv_path: str):
        print("‚öôÔ∏è  Initializing Matching System...")
        self.csv_path = csv_path
        self.df = self._load_dataset()

        print("‚è≥ Loading SBERT Model (all-MiniLM-L6-v2)...")
        self.model = SentenceTransformer('all-MiniLM-L6-v2')

        # Pre-compute embeddings for internships to speed up matching
        if self.df is not None:
            self.df['combined_text'] = self.df['position'] + " " + self.df['description'] + " " + self.df['required_skills']
            self.internship_embeddings = self.model.encode(self.df['combined_text'].tolist(), convert_to_tensor=True)
            print("‚úÖ Model loaded and Internship Embeddings computed.")

    def _load_dataset(self):
        if not os.path.exists(self.csv_path):
            print(f"‚ùå Dataset not found at {self.csv_path}")
            return None
        df = pd.read_csv(self.csv_path)
        # Handle NaN values
        df.fillna('', inplace=True)
        print(f"‚úÖ Loaded {len(df)} internships.")
        return df

    def calculate_ats_score(self, cv_data: Dict) -> Dict:
        """Calculates completeness score"""
        score = 0
        breakdown = {}

        # Personal Info (20 pts)
        p_score = 0
        if cv_data['personal_info']['name'] != "Not Found": p_score += 5
        if cv_data['personal_info']['email'] != "Not Found": p_score += 5
        if cv_data['personal_info']['phone'] != "Not Found": p_score += 5
        if cv_data['personal_info']['website'] != "Not Found": p_score += 5
        score += p_score
        breakdown['Personal Info'] = f"{p_score}/20"

        # Education (30 pts)
        e_score = min(30, len(cv_data['education']) * 15)
        score += e_score
        breakdown['Education'] = f"{e_score}/30"

        # Experience (30 pts)
        ex_score = min(30, len(cv_data['experience']) * 10)
        score += ex_score
        breakdown['Experience'] = f"{ex_score}/30"

        # Skills (20 pts)
        s_score = min(20, len(cv_data['skills']) * 2)
        score += s_score
        breakdown['Skills'] = f"{s_score}/20"

        return {"total": score, "breakdown": breakdown}

    def match_student(self, cv_data: Dict, top_n: int = 5):
        if self.df is None: return None

        # Create student embedding
        student_text = " ".join(cv_data['skills']) + " " + " ".join(cv_data['experience']) + " " + " ".join(cv_data['education'])
        student_embedding = self.model.encode(student_text, convert_to_tensor=True)

        # Calculate Similarity
        cos_scores = util.cos_sim(student_embedding, self.internship_embeddings)[0]

        # Get ATS Score for weighting
        ats_result = self.calculate_ats_score(cv_data)
        ats_weight = ats_result['total'] / 100.0

        results = []
        for idx, score in enumerate(cos_scores):
            # Calculate Skill Overlap Percentage
            req_skills = self.df.iloc[idx]['required_skills'].split(',')
            req_skills = [s.strip().lower() for s in req_skills if s]

            student_skills_lower = [s.lower() for s in cv_data['skills']]

            if req_skills:
                overlap = len(set(req_skills).intersection(student_skills_lower))
                skill_match = overlap / len(req_skills)
            else:
                skill_match = 0

            # Final Score Formula: 40% Semantic + 40% Skills + 20% ATS
            final_score = (float(score) * 0.4) + (skill_match * 0.4) + (ats_weight * 0.2)

            results.append({
                "company": self.df.iloc[idx]['company'],
                "position": self.df.iloc[idx]['position'],
                "location": self.df.iloc[idx]['location'],
                "stipend": self.df.iloc[idx]['monthly_stipend'],
                "score": round(final_score * 100, 2)
            })

        # Sort and return top N
        results = sorted(results, key=lambda x: x['score'], reverse=True)[:top_n]
        return results, ats_result


# ==========================================
# 4. MAIN EXECUTION
# ==========================================
def main():
    # Setup
    cv_file = "Anas_CV.pdf"
    csv_file = "palestinian_internships_200.csv"

    # Step 1: Initialize System
    extractor = CVExtractor()
    matcher = InternshipMatchingSystem(csv_file)

    # Step 2: Extract & Save JSON
    cv_data = extractor.process_cv(cv_file)

    # Step 3: Match & Score
    print("\nü§ñ performing AI Matching...")
    matches, ats_stats = matcher.match_student(cv_data)

    # Step 4: Visualizing Results
    print("\n" + "="*50)
    print("üìä FINAL ANALYSIS REPORT")
    print("="*50)

    # Print ATS Score
    score = ats_stats['total']
    grade = "A" if score >= 80 else "B" if score >= 60 else "C"
    print(f"\nüéØ ATS SCORE: {score}/100 (Grade: {grade})")
    for k, v in ats_stats['breakdown'].items():
        print(f"  - {k}: {v}")

    # Print Matches
    print(f"\nüèÜ TOP {len(matches)} INTERNSHIP RECOMMENDATIONS:")
    print(f"{'#':<3} {'Company':<20} {'Position':<25} {'Loc':<10} {'Stipend':<8} {'Score':<5}")
    print("-" * 80)

    companies = []
    scores = []
    labels = []

    for i, m in enumerate(matches, 1):
        print(f"{i:<3} {m['company']:<20} {m['position']:<25} {m['location']:<10} ${m['stipend']:<7} {m['score']}%")
        companies.append(m['company'])
        scores.append(m['score'])
        labels.append(f"{m['company']}\n({m['position']})")

    # Step 5: Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(x=scores, y=labels, palette="viridis")
    plt.title(f"Top Internship Matches for {cv_data['personal_info']['name']}")
    plt.xlabel("Match Score (%)")
    plt.xlim(0, 100)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

# More Advanced Plotting & Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi

class DashboardVisualizer:
    """
    Advanced Visualization Module for the Palestinian Internship Matching System.
    Generates a 4-panel dashboard.
    """

    def __init__(self):
        # Set a professional style
        sns.set_theme(style="whitegrid")
        self.colors = sns.color_palette("viridis", 10)

    def plot_dashboard(self, cv_data: dict, matches: list, ats_stats: dict):
        """
        Main function to generate the 4-panel dashboard
        """
        student_name = cv_data['personal_info']['name']

        # Create a figure with 2x2 subplots
        fig = plt.figure(figsize=(18, 12))
        fig.suptitle(f" Internship Analysis Dashboard: {student_name}", fontsize=20, fontweight='bold', y=0.95)

        # 1. RADAR CHART (ATS Profile)
        self._plot_radar_chart(fig, ats_stats, 221)

        # 2. MATCH SCORE BAR CHART
        self._plot_match_scores(fig, matches, 222)

        # 3. STIPEND VS MATCH SCORE SCATTER
        self._plot_stipend_vs_score(fig, matches, 223)

        # 4. LOCATION DISTRIBUTION
        self._plot_location_pie(fig, matches, 224)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

    def _plot_radar_chart(self, fig, ats_stats, position):
        """
        Panel 1: Radar Chart for ATS Breakdown
        """
        ax = fig.add_subplot(position, polar=True)

        # Parse data from format "15/20" to integers
        categories = []
        values = []
        max_values = []

        for k, v in ats_stats['breakdown'].items():
            score, max_val = v.split('/')
            categories.append(k)
            values.append(int(score))
            max_values.append(int(max_val))

        # Normalize values to 0-100 scale for the shape to look right
        normalized_values = [(v/m)*100 for v, m in zip(values, max_values)]

        # Close the loop for the radar chart
        N = len(categories)
        angles = [n / float(N) * 2 * pi for n in range(N)]
        normalized_values += [normalized_values[0]]
        angles += [angles[0]]

        # Plot
        ax.plot(angles, normalized_values, linewidth=2, linestyle='solid', color='blue')
        ax.fill(angles, normalized_values, 'blue', alpha=0.1)

        # Fix labels
        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(categories, fontsize=10, fontweight='bold')

        # Set Title
        ax.set_title("ATS Profile Strength", size=14, pad=20)
        ax.set_yticks([25, 50, 75, 100])
        ax.set_yticklabels(["25%", "50%", "75%", "100%"], color="grey", size=8)
        ax.set_ylim(0, 100)

    def _plot_match_scores(self, fig, matches, position):
        """
        Panel 2: Horizontal Bar Chart of Top Matches
        """
        ax = fig.add_subplot(position)

        companies = [m['company'] for m in matches]
        scores = [m['score'] for m in matches]

        # Create Bar Plot
        sns.barplot(x=scores, y=companies, palette="magma", ax=ax, orient='h')

        # Add text labels on bars
        for i, v in enumerate(scores):
            ax.text(v + 1, i, f"{v}%", color='black', va='center', fontweight='bold')

        ax.set_xlim(0, 110)
        ax.set_title("Top Internship Matches", size=14)
        ax.set_xlabel("Match Probability (%)")
        ax.set_ylabel("")

    def _plot_stipend_vs_score(self, fig, matches, position):
        """
        Panel 3: Scatter Plot (ROI Analysis - Stipend vs Fit)
        """
        ax = fig.add_subplot(position)

        scores = [m['score'] for m in matches]
        stipends = [m['stipend'] for m in matches]
        companies = [m['company'] for m in matches]

        # Scatter plot
        sns.scatterplot(x=stipends, y=scores, hue=companies, s=200, style=companies, palette="deep", ax=ax)

        # Add faint grid
        ax.grid(True, linestyle='--', alpha=0.6)

        # Labels
        ax.set_title("Value Analysis: Stipend vs. Match Fit", size=14)
        ax.set_xlabel("Monthly Stipend ($)")
        ax.set_ylabel("Match Score (%)")
        ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

    def _plot_location_pie(self, fig, matches, position):
        """
        Panel 4: Donut Chart for Location Distribution
        """
        ax = fig.add_subplot(position)

        # Count locations
        locations = [m['location'] for m in matches]
        unique_locs = list(set(locations))
        counts = [locations.count(l) for l in unique_locs]

        # Donut Chart
        wedges, texts, autotexts = ax.pie(counts, labels=unique_locs, autopct='%1.1f%%',
                                          startangle=90, colors=sns.color_palette("pastel"),
                                          pctdistance=0.85, wedgeprops=dict(width=0.3))

        # Style text
        for text in texts:
            text.set_fontsize(10)
            text.set_fontweight('bold')

        # Add center text
        center_circle = plt.Circle((0,0),0.70,fc='white')

        ax.set_title("Opportunity Locations", size=14)
```

```python
def main():
    # 1. Setup & Initialization
    cv_file = "Anas_CV.pdf"
    csv_file = "palestinian_internships_200.csv"

    print("üöÄ STARTING PALESTINIAN INTERNSHIP MATCHING SYSTEM...\n")

    extractor = CVExtractor()
    matcher = InternshipMatchingSystem(csv_file)
    visualizer = DashboardVisualizer()  # <--- Initialize the new class

    # 2. Extract CV
    cv_data = extractor.process_cv(cv_file)

    # 3. Match
    print("\nü§ñ Running AI Matching Engine...")
    matches, ats_stats = matcher.match_student(cv_data, top_n=6) # Get top 6 for better plots

    # 4. Text Output
    print("\n" + "="*50)
    print("üìä TEXT SUMMARY")
    print("="*50)
    print(f"Top Recommendation: {matches[0]['position']} at {matches[0]['company']}")
    print(f"Match Score: {matches[0]['score']}%")
    print(f"Stipend: ${matches[0]['stipend']}")

    # 5. Generate Dashboard
    print("\nüé® Generating Advanced Dashboard...")
    visualizer.plot_dashboard(cv_data, matches, ats_stats)

if __name__ == "__main__":
    main()
```

```python
import pdfplumber
import spacy
import re
import phonenumbers
import json
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List
from sentence_transformers import SentenceTransformer, util
from reportlab.pdfgen import canvas  # Only used to create the dummy PDF for testing

# ==========================================
# 1. THE CV EXTRACTOR CLASS (ATS SYSTEM)
# ==========================================
class CVExtractor:
    """
    Extracts text from PDF, parses it into structured data,
    prints the result, and saves to JSON.
    """
    def __init__(self):
        print("‚öôÔ∏è  Initializing CV Extractor (Loading NLP models)...")
        try:
            self.nlp = spacy.load("en_core_web_sm")
        except:
            from spacy.cli import download
            download("en_core_web_sm")
            self.nlp = spacy.load("en_core_web_sm")

    def process_cv(self, pdf_path: str) -> Dict:
        """
        Main function to Extract -> Print -> Save
        """
        print(f"\nüìÇ Reading PDF: {pdf_path}...")
        raw_text = self._extract_text_from_pdf(pdf_path)

        # Parse data
        extracted_data = {
            "personal_info": self._extract_personal_info(raw_text),
            "education": self._extract_section(raw_text, ["EDUCATION", "ACADEMIC BACKGROUND"]),
            "experience": self._extract_experience_robust(raw_text),
            "skills": self._extract_skills(raw_text),
            "raw_text": raw_text[:500] + "..." # Store snippet only to keep JSON clean
        }

        # 1. PRINT RESULTS
        self._print_extraction_results(extracted_data)

        # 2. SAVE TO JSON
        json_filename = pdf_path.replace(".pdf", "_extracted.json")
        with open(json_filename, 'w', encoding='utf-8') as f:
            json.dump(extracted_data, f, indent=4)
        print(f"\nüíæ Data successfully saved to: {json_filename}")

        return extracted_data

    def _extract_text_from_pdf(self, pdf_path: str) -> str:
        text = ""
        try:
            with pdfplumber.open(pdf_path) as pdf:
                for page in pdf.pages:
                    text += (page.extract_text() or "") + "\n"
        except Exception as e:
            print(f"‚ùå Error reading PDF: {e}")
        return text

    def _extract_personal_info(self, text: str) -> Dict:
        info = {"name": "Not Found", "email": "Not Found", "phone": "Not Found", "website": "Not Found"}

        # Extract Email
        email = re.search(r'[\w\.-]+@[\w\.-]+\.\w+', text)
        if email: info['email'] = email.group(0)

        # Extract Phone (Palestinian format +970/972)
        try:
            for match in phonenumbers.PhoneNumberMatcher(text, "PS"):
                info["phone"] = phonenumbers.format_number(match.number, phonenumbers.PhoneNumberFormat.INTERNATIONAL)
                break
        except: pass

        # Extract Website
        url = re.search(r'https?://[^\s]+', text)
        if url: info['website'] = url.group(0)

        # Extract Name (Heuristic: First non-empty line)
        lines = [l.strip() for l in text.split('\n') if l.strip()]
        if lines: info['name'] = lines[0]

        return info

    def _extract_section(self, text: str, headers: List[str]) -> List[str]:
        """Generic section extractor"""
        lines = text.split('\n')
        content = []
        capture = False

        for line in lines:
            line = line.strip()
            # Start capturing if header matches
            if any(h in line.upper() for h in headers):
                capture = True
                continue

            # Stop capturing if new section starts (common headers)
            common_headers = ['EXPERIENCE', 'EDUCATION', 'SKILLS', 'PROJECTS', 'SUMMARY', 'LANGUAGES']
            if capture and any(h in line.upper() for h in common_headers) and len(line) < 20:
                capture = False

            if capture and line:
                content.append(line)
        return content

    def _extract_experience_robust(self, text: str) -> List[str]:
        """
        Improved Experience Logic:
        1. Extract the experience block.
        2. Use Regex to find dates (years) to identify distinct entries.
        """
        raw_lines = self._extract_section(text, ["EXPERIENCE", "WORK HISTORY", "PROFESSIONAL EXPERIENCE"])
        if not raw_lines: return []

        entries = []
        current_entry = ""

        # Regex for dates like "2021", "2016 - 2018", "Sep 2021"
        date_pattern = re.compile(r'(19|20)\d{2}')

        for line in raw_lines:
            # If a line contains a date and we already have text, push the previous entry
            if date_pattern.search(line) and len(current_entry) > 10:
                entries.append(current_entry.strip())
                current_entry = line # Start new entry
            else:
                current_entry += " " + line

        if current_entry: entries.append(current_entry.strip())

        return entries

    def _extract_skills(self, text: str) -> List[str]:
        raw_lines = self._extract_section(text, ["SKILLS", "TECHNICAL SKILLS"])
        full_text = " ".join(raw_lines)
        # Split by bullets, commas, or newlines
        skills = re.split(r'[,‚Ä¢\n]', full_text)
        # Clean and deduplicate
        return list(set([s.strip() for s in skills if len(s.strip()) > 1]))

    def _print_extraction_results(self, data: Dict):
        print("\n" + "="*50)
        print("üìÑ EXTRACTED CV DATA")
        print("="*50)
        print(f"üë§ Name:    {data['personal_info']['name']}")
        print(f"üìß Email:   {data['personal_info']['email']}")
        print(f"üì± Phone:   {data['personal_info']['phone']}")
        print(f"üåê Website: {data['personal_info']['website']}")

        print("\nüéì Education:")
        for edu in data['education']: print(f"  - {edu}")

        print("\nüíº Experience:")
        for exp in data['experience']: print(f"  - {exp}")

        print("\nüõ†Ô∏è  Skills:")
        print(f"  {', '.join(data['skills'])}")
        print("="*50)


# ==========================================
# 2. THE MATCHING SYSTEM (AI & ATS SCORING)
# ==========================================
class InternshipMatchingSystem:
    def __init__(self, csv_path: str):
        print("‚öôÔ∏è  Initializing Matching System...")
        self.csv_path = csv_path
        self.df = self._load_dataset()

        print("‚è≥ Loading SBERT Model (all-MiniLM-L6-v2)...")
        self.model = SentenceTransformer('all-MiniLM-L6-v2')

        # Pre-compute embeddings for internships to speed up matching
        if self.df is not None:
            self.df['combined_text'] = self.df['position'] + " " + self.df['description'] + " " + self.df['required_skills']
            self.internship_embeddings = self.model.encode(self.df['combined_text'].tolist(), convert_to_tensor=True)
            print("‚úÖ Model loaded and Internship Embeddings computed.")

    def _load_dataset(self):
        if not os.path.exists(self.csv_path):
            print(f"‚ùå Dataset not found at {self.csv_path}")
            return None
        df = pd.read_csv(self.csv_path)
        # Handle NaN values
        df.fillna('', inplace=True)
        print(f"‚úÖ Loaded {len(df)} internships.")
        return df

    def calculate_ats_score(self, cv_data: Dict) -> Dict:
        """Calculates completeness score"""
        score = 0
        breakdown = {}

        # Personal Info (20 pts)
        p_score = 0
        if cv_data['personal_info']['name'] != "Not Found": p_score += 5
        if cv_data['personal_info']['email'] != "Not Found": p_score += 5
        if cv_data['personal_info']['phone'] != "Not Found": p_score += 5
        if cv_data['personal_info']['website'] != "Not Found": p_score += 5
        score += p_score
        breakdown['Personal Info'] = f"{p_score}/20"

        # Education (30 pts)
        e_score = min(30, len(cv_data['education']) * 15)
        score += e_score
        breakdown['Education'] = f"{e_score}/30"

        # Experience (30 pts)
        ex_score = min(30, len(cv_data['experience']) * 10)
        score += ex_score
        breakdown['Experience'] = f"{ex_score}/30"

        # Skills (20 pts)
        s_score = min(20, len(cv_data['skills']) * 2)
        score += s_score
        breakdown['Skills'] = f"{s_score}/20"

        return {"total": score, "breakdown": breakdown}

    def match_student(self, cv_data: Dict, top_n: int = 5):
        if self.df is None: return None

        # Create student embedding
        student_text = " ".join(cv_data['skills']) + " " + " ".join(cv_data['experience']) + " " + " ".join(cv_data['education'])
        student_embedding = self.model.encode(student_text, convert_to_tensor=True)

        # Calculate Similarity
        cos_scores = util.cos_sim(student_embedding, self.internship_embeddings)[0]

        # Get ATS Score for weighting
        ats_result = self.calculate_ats_score(cv_data)
        ats_weight = ats_result['total'] / 100.0

        results = []
        for idx, score in enumerate(cos_scores):
            # Calculate Skill Overlap Percentage
            req_skills = self.df.iloc[idx]['required_skills'].split(',')
            req_skills = [s.strip().lower() for s in req_skills if s]

            student_skills_lower = [s.lower() for s in cv_data['skills']]

            if req_skills:
                overlap = len(set(req_skills).intersection(student_skills_lower))
                skill_match = overlap / len(req_skills)
            else:
                skill_match = 0

            # Final Score Formula: 40% Semantic + 40% Skills + 20% ATS
            final_score = (float(score) * 0.4) + (skill_match * 0.4) + (ats_weight * 0.2)

            results.append({
                "company": self.df.iloc[idx]['company'],
                "position": self.df.iloc[idx]['position'],
                "location": self.df.iloc[idx]['location'],
                "stipend": self.df.iloc[idx]['monthly_stipend'],
                "score": round(final_score * 100, 2)
            })

        # Sort and return top N
        results = sorted(results, key=lambda x: x['score'], reverse=True)[:top_n]
        return results, ats_result


# ==========================================
# 4. MAIN EXECUTION
# ==========================================
def main():
    # Setup
    cv_file = "2.pdf"
    csv_file = "palestinian_internships_200.csv"

    # Step 1: Initialize System
    extractor = CVExtractor()
    matcher = InternshipMatchingSystem(csv_file)

    # Step 2: Extract & Save JSON
    cv_data = extractor.process_cv(cv_file)

    # Step 3: Match & Score
    print("\nü§ñ performing AI Matching...")
    matches, ats_stats = matcher.match_student(cv_data)

    # Step 4: Visualizing Results
    print("\n" + "="*50)
    print("üìä FINAL ANALYSIS REPORT")
    print("="*50)

    # Print ATS Score
    score = ats_stats['total']
    grade = "A" if score >= 80 else "B" if score >= 60 else "C"
    print(f"\nüéØ ATS SCORE: {score}/100 (Grade: {grade})")
    for k, v in ats_stats['breakdown'].items():
        print(f"  - {k}: {v}")

    # Print Matches
    print(f"\nüèÜ TOP {len(matches)} INTERNSHIP RECOMMENDATIONS:")
    print(f"{'#':<3} {'Company':<20} {'Position':<25} {'Loc':<10} {'Stipend':<8} {'Score':<5}")
    print("-" * 80)

    companies = []
    scores = []
    labels = []

    for i, m in enumerate(matches, 1):
        print(f"{i:<3} {m['company']:<20} {m['position']:<25} {m['location']:<10} ${m['stipend']:<7} {m['score']}%")
        companies.append(m['company'])
        scores.append(m['score'])
        labels.append(f"{m['company']}\n({m['position']})")

    # Step 5: Plotting
    plt.figure(figsize=(10, 6))
    sns.barplot(x=scores, y=labels, palette="viridis")
    plt.title(f"Top Internship Matches for {cv_data['personal_info']['name']}")
    plt.xlabel("Match Score (%)")
    plt.xlim(0, 100)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
