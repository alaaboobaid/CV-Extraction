```python
"""
AI CV System: Full Verbose Mode, Local Model Storage, and Advanced Analytics
Version: 10.0 (Ultimate Edition)
"""

import os
import re
import json
import warnings
import subprocess
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pdfplumber
import spacy
import phonenumbers
from sentence_transformers import SentenceTransformer, util

# Configuration
warnings.filterwarnings('ignore')
MODEL_DIR = "cv_ai_models"
OUTPUT_DIR = "output"
os.makedirs(MODEL_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ==========================================
# 1. MODEL PERSISTENCE (SAVE TO DISK)
# ==========================================
def initialize_local_models():
    print("\n" + "="*60)
    print("📂 MANAGING LOCAL AI MODELS")
    print("="*60)

    spacy_local_path = os.path.join(MODEL_DIR, "en_core_web_sm_local")
    if not os.path.exists(spacy_local_path):
        print("📥 Downloading spaCy (en_core_web_sm)...")
        subprocess.check_call([sys.executable, "-m", "spacy", "download", "en_core_web_sm"])
        nlp = spacy.load("en_core_web_sm")
        nlp.to_disk(spacy_local_path)
    else:
        print(f"✅ Loaded spaCy from local drive: {spacy_local_path}")

    sbert_local_path = os.path.join(MODEL_DIR, "sbert_transformer_local")
    if not os.path.exists(sbert_local_path):
        print("📥 Downloading SBERT (all-MiniLM-L6-v2)...")
        model = SentenceTransformer('all-MiniLM-L6-v2')
        model.save(sbert_local_path)
    else:
        print(f"✅ Loaded SBERT from local drive: {sbert_local_path}")

    return spacy_local_path, sbert_local_path

# ==========================================
# 2. THE EXTRACTION ENGINE (VERBOSE)
# ==========================================
class EliteCVEngine:
    def __init__(self, spacy_path):
        self.nlp = spacy.load(spacy_path)
        self.headers = {
            'education': ['EDUCATION', 'ACADEMIC', 'QUALIFICATIONS', 'UNIVERSITY', 'DEGREES'],
            'experience': ['EXPERIENCE', 'EMPLOYMENT', 'WORK HISTORY', 'PROFESSIONAL'],
            'skills': ['SKILLS', 'TECHNICAL SKILLS', 'TECHNOLOGIES', 'COMPETENCIES'],
            'projects': ['PROJECTS', 'KEY PROJECTS', 'PORTFOLIO']
        }

    def process(self, pdf_path):
        print("\n" + "="*60)
        print(f"📄 EXTRACTING DATA FROM: {pdf_path}")
        print("="*60)

        lines = []
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                text = page.extract_text()
                if text: lines.extend(text.split('\n'))

        # Strict Section Segmentation
        sections = {k: [] for k in self.headers.keys()}
        current_sec = None

        for line in lines:
            clean = line.strip().upper()
            if not clean: continue

            is_header = False
            for sec, keys in self.headers.items():
                if any(clean == k or clean.startswith(k) for k in keys) and len(clean) < 40:
                    current_sec = sec
                    is_header = True
                    break

            if is_header: continue
            if current_sec:
                sections[current_sec].append(line.strip())

        # Skill Refinement using NLP Part-of-Speech Tagging
        raw_skill_text = " ".join(sections['skills']).replace(',', ' ')
        tokens = self.nlp(raw_skill_text)
        refined_skills = list(dict.fromkeys([
            t.text.capitalize() for t in tokens
            if (t.pos_ in ['PROPN', 'NOUN'] and len(t.text) > 2)
        ]))

        full_text = "\n".join(lines)

        # Identity Extraction
        phone = "Not Found"
        for m in phonenumbers.PhoneNumberMatcher(full_text, "PS"):
            phone = phonenumbers.format_number(m.number, phonenumbers.PhoneNumberFormat.INTERNATIONAL)

        email = "Not Found"
        e_match = re.search(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', full_text)
        if e_match: email = e_match.group(0)

        name = lines[0]
        for l in lines[:5]:
            doc = self.nlp(l)
            for ent in doc.ents:
                if ent.label_ == "PERSON":
                    name = ent.text
                    break

        data = {
            "personal": {"name": name, "phone": phone, "email": email},
            "education": list(dict.fromkeys(sections['education'])),
            "experience": sections['experience'],
            "projects": sections['projects'],
            "skills": refined_skills
        }

        # PRINT ALL EXTRACTED DATA
        print(json.dumps(data, indent=4))
        return data

# ==========================================
# 3. ANALYTICS & DASHBOARD
# ==========================================
def run_matching_and_viz(cv_data, csv_path, sbert_path):
    print("\n" + "="*60)
    print("📈 SCORING & SEMANTIC MATCHING")
    print("="*60)

    # 1. Load Matcher
    model = SentenceTransformer(sbert_path)
    df = pd.read_csv(csv_path).fillna('')

    # 2. Calculate ATS Score
    # Experience (45 pts), Skills (30 pts), Education (15 pts), Contact (10 pts)
    # Projects are automatically included in Experience logic
    exp_count = len(cv_data['experience']) + len(cv_data['projects'])
    score_details = {
        "Experience": min(45, exp_count * 6),
        "Skills": min(30, len(cv_data['skills']) * 3),
        "Education": min(15, len(cv_data['education']) * 5),
        "Contact": 10 if cv_data['personal']['phone'] != "Not Found" else 5
    }
    ats_total = sum(score_details.values())

    print(f"ATS SCORING BREAKDOWN:")
    for k, v in score_details.items():
        print(f" - {k}: {v} pts")
    print(f"TOTAL ATS SCORE: {ats_total}/100\n")

    # 3. Semantic Matching
    cv_context = f"{' '.join(cv_data['skills'])} {' '.join(cv_data['experience'])} {' '.join(cv_data['projects'])}"
    cv_emb = model.encode(cv_context, convert_to_tensor=True)
    db_emb = model.encode((df['position'] + " " + df['description']).tolist(), convert_to_tensor=True)

    similarities = util.cos_sim(cv_emb, db_emb)[0]

    matches = []
    for i, sim in enumerate(similarities):
        # 70% Semantic Text Matching + 30% Profile Strength (ATS)
        final = (float(sim) * 0.7) + (ats_total/100 * 0.3)
        matches.append({"company": df.iloc[i]['company'], "position": df.iloc[i]['position'], "score": round(final * 100, 2)})

    matches = sorted(matches, key=lambda x: x['score'], reverse=True)

    print(f"TOP 10 INTERNSHIP MATCHES:")
    for i, m in enumerate(matches[:10], 1):
        print(f" {i}. {m['company']} | {m['position']} | {m['score']}% Match")

    # --- 4. VISUALIZATION ---
    print("\n🎨 Saving Dashboard to /output...")
    sns.set_theme(style="white")
    fig = plt.figure(figsize=(20, 14))

    # Chart 1: Donut (Score Breakdown)
    ax1 = fig.add_subplot(231)
    ax1.pie(score_details.values(), labels=score_details.keys(), autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"), wedgeprops=dict(width=0.4))
    ax1.set_title("ATS Profile Composition", fontweight='bold')

    # Chart 2: Rose Chart (Readiness)
    ax2 = fig.add_subplot(232, polar=True)
    max_vals = [45, 30, 15, 10]
    norm = [(v/m)*100 for v, m in zip(score_details.values(), max_vals)]
    angles = np.linspace(0, 2*np.pi, len(norm), endpoint=False)
    ax2.bar(angles, norm, width=1.0, color=sns.color_palette("viridis", len(norm)), alpha=0.6)
    ax2.set_xticks(angles)
    ax2.set_xticklabels(score_details.keys())
    ax2.set_title("Readiness Level (%)", pad=20, fontweight='bold')

    # Chart 3: Top Skills
    ax3 = fig.add_subplot(233)
    top_skills = cv_data['skills'][:10]
    sns.barplot(x=list(range(len(top_skills), 0, -1)), y=top_skills, palette="flare", ax=ax3)
    ax3.set_title("Top Skills Extracted", fontweight='bold')

    # Chart 4: Match Distribution
    ax4 = fig.add_subplot(234)
    top_5 = matches[:5]
    ax4.pie([m['score'] for m in top_5], labels=[m['company'] for m in top_5], autopct='%1.1f%%', colors=sns.color_palette("mako"))
    ax4.set_title("Top 5 Match Distribution", fontweight='bold')

    # Chart 5: Global Ranking
    ax5 = fig.add_subplot(2, 3, (5, 6))
    top_df = pd.DataFrame(matches).head(10)
    sns.barplot(data=top_df, x='score', y='company', palette="viridis", ax=ax5)
    ax5.set_title("Top 10 Company Match Confidence", fontweight='bold')
    ax5.set_xlim(0, 100)

    plt.tight_layout()
    plt.savefig(os.path.join(OUTPUT_DIR, "executive_summary.png"), dpi=300)
    plt.show()

# ==========================================
# MAIN LOOP
# ==========================================
if __name__ == "__main__":
    CV_PATH = "Anas_CV.pdf"
    CSV_PATH = "palestinian_internships_200.csv"

    if os.path.exists(CV_PATH) and os.path.exists(CSV_PATH):
        # Initializing Models
        SPACY_PATH, SBERT_PATH = initialize_local_models()

        # Running Extraction
        engine = EliteCVEngine(SPACY_PATH)
        cv_data = engine.process(CV_PATH)

        # Running Analytics
        run_matching_and_viz(cv_data, CSV_PATH, SBERT_PATH)

        # Save JSON
        with open(os.path.join(OUTPUT_DIR, "full_cv_report.json"), "w") as f:
            json.dump(cv_data, f, indent=4)

        print("\n✅ PROCESS COMPLETED SUCCESSFULLY.")
        print(f"📁 JSON and Visual Dashboard saved in the '{OUTPUT_DIR}' folder.")
    else:
        print("❌ ERROR: Ensure 'Anas_CV.pdf' and 'palestinian_internships_200.csv' are in the root folder.")
```
